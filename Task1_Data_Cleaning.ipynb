```python
# ===========================================
# Experiment #01
# Task: Data Cleaning and Preprocessing
# Dataset: Netflix Movies and TV Shows
# ===========================================

# Step 1: Import Libraries
import pandas as pd
import numpy as np

# Step 2: Load Dataset
df = pd.read_csv("/content/netflix.csv")

print("First Five Records")
print(df.head())

# Step 3: Dataset Information
print("\nDataset Shape:", df.shape)

print("\nDataset Information")
print(df.info())

# Step 4: Check Missing Values
print("\nMissing Values")
print(df.isnull().sum())

# Step 5: Check Duplicate Records
print("\nDuplicate Records:", df.duplicated().sum())

# Step 6: Remove Duplicate Records
df = df.drop_duplicates()

print("\nDuplicate Records After Removal:", df.duplicated().sum())

# Step 7: Handle Missing Values

# Fill missing values in text columns
# df["director"] = df["director"].fillna("Unknown") # Column 'director' not found in df.info()
# df["cast"] = df["cast"].fillna("Not Available") # Column 'cast' not found in df.info()
# df["country"] = df["country"].fillna("Unknown") # Column 'country' not found in df.info()

# Fill rating with mode
df["rating"] = df["rating"].fillna(df["rating"].mode()[0])

# Remove rows where date_added is missing
# df = df.dropna(subset=["date_added"]) # Column 'date_added' not found in df.info()

print("\nMissing Values After Cleaning")
print(df.isnull().sum())

# Step 8: Rename Column Names
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")

print("\nUpdated Column Names")
print(df.columns)

# Step 9: Standardize Text Data

# df["type"] = df["type"].str.title() # Column 'type' not found in df.info()

# df["country"] = df["country"].str.title() # Column 'country' not found in df.info()

df["rating"] = df["rating"].str.upper()

# Step 10: Convert Date Format

# df["date_added"] = pd.to_datetime(df["date_added"]) # Column 'date_added' not found in df.info()

# df["date_added"] = df["date_added"].dt.strftime("%d-%m-%Y") # Column 'date_added' not found in df.info()

# Step 11: Check Data Types

print("\nData Types")
print(df.dtypes)

# Step 12: Display Cleaned Dataset

print("\nFirst Five Records After Cleaning")
print(df.head())

# Step 13: Save Cleaned Dataset

df.to_csv("/content/netflix_titles_cleaned.csv", index=False)

print("\nCleaned dataset saved successfully!")

# Step 14: Summary

print("\n========== DATA CLEANING SUMMARY ==========")
print("Original Dataset : Netflix Movies and TV Shows")
print("✔ Missing values handled")
print("✔ Duplicate records removed")
print("✔ Column names standardized")
print("✔ Text values standardized")
print("✔ Date format converted")
print("✔ Data types verified")
print("✔ Cleaned dataset saved as netflix_titles_cleaned.csv")
```
